```python
import pandas as pd
import numpy as np
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN',
'londON_StockhOlm','Budapest_PaRis', 'Brussels_londOn'],
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',
'12. Air France', '"Swiss Air"']})
```

```python
df_temp=df.copy()
df_temp[['From','To']] = df.From_To.apply( 
   lambda x: pd.Series(str(x).split("_"))) 
```

```python
df_temp.head()
```

```python
df_temp['From']=df_temp.From.str.capitalize()
df_temp['To']=df_temp.To.str.capitalize()
```

```python
delays = pd.DataFrame(df.RecentDelays.values.tolist())
delays.columns = ['delay_1', 'delay_2', 'delay_3']
```

```python
df_temp = pd.concat([df_temp, delays[['delay_1', 'delay_2', 'delay_3']]], axis=1)
```
